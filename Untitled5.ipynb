# *Price Distribution*

```python
plt.figure(figsize=(8, 8))
sns.histplot(df['Price'], kde=True, color='blue', bins=10)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(rotation='vertical')
plt.title('Price Distribution of Laptops')
plt.show()
```

# *Price Occurrences in Different Ranges*

```python
df_1 = df[df['Price'] < 30000]
df_2 = df[(df['Price'] >= 30000) & (df['Price'] < 50000)]
df_3 = df[(df['Price'] >= 50000) & (df['Price'] < 70000)]
df_4 = df[(df['Price'] >= 70000) & (df['Price'] < 80000)]
df_5 = df[(df['Price'] >= 80000) & (df['Price'] < 100000)]
df_6 = df[(df['Price'] >= 100000) & (df['Price'] < 300000)]

counts_1 = df_1['Price'].count()
counts_2 = df_2['Price'].count()
counts_3 = df_3['Price'].count()
counts_4 = df_4['Price'].count()
counts_5 = df_5['Price'].count()
counts_6 = df_6['Price'].count()


plt.figure(figsize=(10, 6))
sns.barplot(x=['Less than 30000', '30000 - 50000', '50000 - 70000','70000-80000','80000-100000','100000-300000'], y=[counts_1, counts_2, counts_3,counts_4,counts_5,counts_6], palette='muted')

plt.xlabel('Price Ranges')
plt.ylabel('Number of Laptops')
plt.title('Price Occurrences in Different Ranges')

plt.show()
```

In this bar graph, we have categorized the laptops based on their respective price ranges, and subsequently computed the counts of laptops falling within each category. ***The dataset has been partitioned into six distinct groups, between 30,000 and 50,000, between 50,000 and 70,000, between 70,000 and 80,000, between 80,000 and 100,000, and finally, those priced between 100,000 and 300,000***
From the bar garph, it becomes evident that the most prevalent price range for laptops is between 50,000 to 70,000 . Following this, the next frequently occurring price range lies between 30,000 to 50,000 .

# *Most Sold Laptop Brand On Flipkart*

```python
brand_sales = df.groupby('Company')['Sales in a year'].sum().reset_index()
labels = brand_sales['Company']
sizes = brand_sales['Sales in a year']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.title('Most Sold Laptop Brands')
plt.axis('equal')

brand_sales_sorted=brand_sales.sort_values(by='Sales in a year' , ascending=False, )
print(brand_sales_sorted)

plt.show()


```

This pie chart provides valuable insights into the sales distribution of laptop brands on the Flipkart platform in one year. ***The chart's data reveals that Asus has the highest sale percentage of 27.2 % among all laptop brands on Flipkart, followed by HP, Lenevo, Acer, and so on...***By analyzing this pie chart, viewers can gain a comprehensive understanding of the market share and popularity of various laptop brands on Flipkart. This information is crucial for consumers, as it indicates the leading brands with the highest sales volume.

```python
company_sales = df.groupby('Company')['Sales in a year'].sum().reset_index()

company_sales_sorted = company_sales.sort_values(by='Sales in a year', ascending=False)

companies = company_sales_sorted['Company']
sales = company_sales_sorted['Sales in a year']

plt.figure(figsize=(12, 8))
plt.barh(companies, sales, color='skyblue')

plt.xlabel('Number of Sales')
plt.ylabel('Company Name')
plt.title('Laptop Sales by Company on Flipkart')

plt.tight_layout()
plt.show()
```

The bar graph presents the total number of laptops sold by various companies in a year. ***Asus stands as the top seller with 57,498 laptops sold, closely followed by HP with 47,630 sales. Lenovo secures the third position with 34,227 units sold, while Acer and Apple follow with 23,828 and 11,951 sales, respectively. The graph further showcases the sales figures for Microsoft, Dell, MSI, Alienware, Avita, Nexstgo, and MarQ, revealing their respective contributions to the laptop market.*** This insightful visualization provides a clear overview of the market share distribution, aiding stakeholders in understanding the relative popularity and performance of laptop brands in the given time frame.
